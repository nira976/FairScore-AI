```python
import pandas as pd
import numpy as np

# Number of records
n = 1000

# Reproducibility
np.random.seed(42)

# Generate Features

monthly_income = np.random.randint(10000, 100001, n)

income_stability = np.random.randint(1, 11, n)

savings_ratio = np.round(
    np.random.uniform(0.01, 0.40, n), 2
)

upi_transactions = np.random.randint(20, 501, n)

bill_payment_score = np.random.randint(50, 101, n)

emi_defaults = np.random.randint(0, 6, n)

spending_volatility = np.random.randint(1, 11, n)

employment_duration = np.random.randint(1, 61, n)

existing_debt_ratio = np.round(
    np.random.uniform(0.00, 0.80, n), 2
)

# Risk Classification

risk_category = []

for i in range(n):

    score = 0

    # Monthly Income

    if monthly_income[i] >= 50000:
        score += 2
    elif monthly_income[i] >= 30000:
        score += 1

    # Income Stability

    if income_stability[i] >= 7:
        score += 2
    elif income_stability[i] >= 5:
        score += 1

    # Savings Ratio

    if savings_ratio[i] >= 0.20:
        score += 2
    elif savings_ratio[i] >= 0.10:
        score += 1

    # Bill Payment Score

    if bill_payment_score[i] >= 90:
        score += 2
    elif bill_payment_score[i] >= 75:
        score += 1

    # EMI Defaults

    if emi_defaults[i] == 0:
        score += 2
    elif emi_defaults[i] <= 2:
        score += 1

    # Spending Volatility

    if spending_volatility[i] <= 3:
        score += 2
    elif spending_volatility[i] <= 6:
        score += 1

    # Employment Duration

    if employment_duration[i] >= 24:
        score += 1

    # Existing Debt Ratio

    if existing_debt_ratio[i] <= 0.30:
        score += 2
    elif existing_debt_ratio[i] <= 0.50:
        score += 1

    # Final Risk Category

    if score >= 10:
        risk_category.append("Low Risk")

    elif score >= 6:
        risk_category.append("Moderate Risk")

    else:
        risk_category.append("High Risk")

# Create DataFrame

df = pd.DataFrame({
    'monthly_income': monthly_income,
    'income_stability': income_stability,
    'savings_ratio': savings_ratio,
    'upi_transactions': upi_transactions,
    'bill_payment_score': bill_payment_score,
    'emi_defaults': emi_defaults,
    'spending_volatility': spending_volatility,
    'employment_duration': employment_duration,
    'existing_debt_ratio': existing_debt_ratio,
    'risk_category': risk_category
})

# Display first 5 rows

print("\nFirst 5 Rows:\n")
print(df.head())

# Check Risk Distribution

print("\nRisk Category Distribution:\n")
print(df['risk_category'].value_counts())

# Save Dataset

df.to_csv("gig_worker_credit_data.csv", index=False)

print("\nDataset saved successfully as 'gig_worker_credit_data.csv'")
```

```python
import pandas as pd
import numpy as np

# ==========================================
# FAIRSCORE AI - DATASET GENERATOR V2
# ==========================================

np.random.seed(42)

n = 1000

# ------------------------------------------
# Worker Type
# ------------------------------------------

worker_type = np.random.choice(
    [
        "Delivery Partner",
        "Ride Share Driver",
        "Freelancer",
        "Content Creator"
    ],
    size=n,
    p=[0.35, 0.30, 0.25, 0.10]
)

# ------------------------------------------
# Generate Features
# ------------------------------------------

monthly_income = []
income_stability = []
customer_rating = []

for worker in worker_type:

    if worker == "Delivery Partner":
        monthly_income.append(np.random.randint(18000, 50000))
        income_stability.append(np.random.randint(6, 10))
        customer_rating.append(round(np.random.uniform(3.8, 5.0), 1))

    elif worker == "Ride Share Driver":
        monthly_income.append(np.random.randint(20000, 70000))
        income_stability.append(np.random.randint(5, 9))
        customer_rating.append(round(np.random.uniform(3.5, 5.0), 1))

    elif worker == "Freelancer":
        monthly_income.append(np.random.randint(15000, 120000))
        income_stability.append(np.random.randint(2, 9))
        customer_rating.append(round(np.random.uniform(3.5, 5.0), 1))

    else:  # Content Creator

        monthly_income.append(np.random.randint(10000, 150000))
        income_stability.append(np.random.randint(1, 8))
        customer_rating.append(round(np.random.uniform(3.0, 5.0), 1))

monthly_income = np.array(monthly_income)
income_stability = np.array(income_stability)
customer_rating = np.array(customer_rating)

# ------------------------------------------
# Other Financial Features
# ------------------------------------------

savings_ratio = np.round(
    np.random.uniform(0.01, 0.40, n),
    2
)

upi_transactions = np.random.randint(
    20,
    500,
    n
)

bill_payment_score = np.random.randint(
    50,
    101,
    n
)

emi_defaults = np.random.randint(
    0,
    6,
    n
)

spending_volatility = np.random.randint(
    1,
    11,
    n
)

employment_duration = np.random.randint(
    1,
    61,
    n
)

existing_debt_ratio = np.round(
    np.random.uniform(0.00, 0.80, n),
    2
)

platform_dependency = np.random.randint(
    20,
    101,
    n
)

emergency_fund_months = np.random.randint(
    0,
    13,
    n
)

# ------------------------------------------
# Financial Health Score
# ------------------------------------------

financial_health_score = []

for i in range(n):

    score = 50

    score += savings_ratio[i] * 50

    score -= existing_debt_ratio[i] * 40

    score -= emi_defaults[i] * 5

    score += (bill_payment_score[i] - 50) * 0.5

    score += emergency_fund_months[i] * 2

    score += customer_rating[i] * 2

    score = max(0, min(100, score))

    financial_health_score.append(round(score))

financial_health_score = np.array(
    financial_health_score
)

# ------------------------------------------
# Risk Category
# ------------------------------------------

risk_category = []

for i in range(n):

    score = financial_health_score[i]

    if score >= 75:
        risk_category.append("Low Risk")

    elif score >= 50:
        risk_category.append("Moderate Risk")

    else:
        risk_category.append("High Risk")

# ------------------------------------------
# Create DataFrame
# ------------------------------------------

df = pd.DataFrame({

    "worker_type": worker_type,

    "monthly_income": monthly_income,

    "income_stability": income_stability,

    "customer_rating": customer_rating,

    "savings_ratio": savings_ratio,

    "upi_transactions": upi_transactions,

    "bill_payment_score": bill_payment_score,

    "emi_defaults": emi_defaults,

    "spending_volatility": spending_volatility,

    "employment_duration": employment_duration,

    "existing_debt_ratio": existing_debt_ratio,

    "platform_dependency": platform_dependency,

    "emergency_fund_months": emergency_fund_months,

    "financial_health_score": financial_health_score,

    "risk_category": risk_category

})

# ------------------------------------------
# Save Dataset
# ------------------------------------------

df.to_csv(
    "gig_worker_credit_data_v2.csv",
    index=False
)

print("Dataset saved successfully!")

print("\nDataset Shape:")
print(df.shape)

print("\nRisk Distribution:")
print(df["risk_category"].value_counts())

print("\nFirst 5 Rows:")
print(df.head())
```
